### Plot all cycleways as of 2020

```python
gdf_freguesias = gpd.read_file("Data/Lisboa_Freguesias/Lisboa_Freguesias_CAOP2015_TM06.shp")
gdf_freguesias = gdf_freguesias.to_crs(epsg=4326)
gdf_freguesias.geometry.index = gdf_freguesias['DICOFRE']

style_1 = {'fillColor': '#1ED674', 'color': '#1ED674'}
style_2 = {'fillColor': '#D6A86E', 'color': '#D6A86E'}
style_3 = {'fillColor': '#233BF7', 'color': '#233BF7'}

unique_zones = gdf_freguesias['DICOFRE'].unique()
m = folium.Map([38.748662, -9.145801],
                 zoom_start=12,
                 tiles='cartodbpositron')

for i in range(len(ciclovias['features'])):
    ciclovia = ciclovias['features'][i]
    
    try:
        ano = int(ciclovia['properties']['ANO'])
        if ano <= 2020:

            folium.GeoJson(
                ciclovia,
                name='geojson',
                style_function=lambda x:style_1
            ).add_to(m)
    except ValueError:
        folium.GeoJson(
                ciclovia,
                name='geojson',
                style_function=lambda x:style_3
            ).add_to(m)
    
for i, zone in enumerate(unique_zones):
    mask = gdf_freguesias['DICOFRE'] == zone

    example_or = gdf_freguesias.loc[mask]
    folium.GeoJson(example_or,name='polygon_or',style_function=lambda x:style_2).add_to(m)

m
```

### Create maps of Lisbon's cycleways over time

```python
gdf_freguesias = gpd.read_file("Data/Lisboa_Freguesias/Lisboa_Freguesias_CAOP2015_TM06.shp")
gdf_freguesias = gdf_freguesias.to_crs(epsg=4326)
gdf_freguesias.geometry.index = gdf_freguesias['DICOFRE']
style_0 = {'fillColor': '#D6A86E', 'color': '#D6A86E'}
style_1 = {'fillColor': '#1ED674', 'color': '#1ED674'}
style_2 = {'fillColor': '#5CC1F7', 'color': '#5CC1F7'}
style_3 = {'fillColor': '#472EAB', 'color': '#472EAB'}
style_4 = {'fillColor': '#AB2E6E', 'color': '#AB2E6E'}

unique_zones = gdf_freguesias['DICOFRE'].unique()


for current_year in range(2013, 2021):

    m = folium.Map([38.747156, -9.156788],
                 zoom_start=13,
                 tiles='cartodbpositron')
    
    for i in range(len(ciclovias['features'])):
        ciclovia = ciclovias['features'][i]

        try:
            ano = int(ciclovia['properties']['ANO'])
            if ano <= current_year:
                folium.GeoJson(
                    ciclovia,
                    name='geojson',
                    style_function=lambda x:style_3
                ).add_to(m)

        except ValueError:
            pass
    
    m.save(str(current_year)+".html")

```
